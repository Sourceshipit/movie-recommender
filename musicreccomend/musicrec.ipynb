```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

df = pd.read_csv('D:\Python mini project\musicreccomend\spotify_millsongdata.csv')
print("Dataset loaded successfully ‚úÖ")
print(df.head())

print("Columns:", df.columns)


df = df.dropna(subset=['text'])
df = df.reset_index(drop=True)

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['text'])

print("TF-IDF shape:", tfidf_matrix.shape)

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

def recommend(song_title, top_n=5):

    idx = df[df['song'].str.lower() == song_title.lower()].index
    if len(idx) == 0:
        print(f"‚ùå '{song_title}' not found in dataset.")
        return
    
    idx = idx[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:top_n+1]
    song_indices = [i[0] for i in sim_scores]
    
    print(f"\nüéß Top {top_n} recommendations for '{song_title}':")
    for i, song_idx in enumerate(song_indices, 1):
        print(f"{i}. {df.iloc[song_idx]['song']} ‚Äî {df.iloc[song_idx]['artist']}")

recommend("HMy Kind Of Girl")
```
